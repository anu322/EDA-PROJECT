## Dataset Details:
- heartRate: Heart rate readings from the smartwatch sensors. -  ppgGreen, ppgRed, ppgIR: PPG (Photoplethysmography) sensor readings in green, 
red, and infrared wavelengths respective.
- drowsiness: label indicating the level of drowsiness based on an adapted karolinska sleepiness scale (kss).values ranges from 0.0 to 0.2 where 0.0
represents alertness and 2.0 represents significant drowsiness..ss. S).
